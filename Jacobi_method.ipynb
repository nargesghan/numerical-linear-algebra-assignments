```python
from pprint import pprint
from numpy import array, zeros, diag, diagflat, dot,set_printoptions,linalg,zeros_like,fill_diagonal,any
import ast
import pprint

def jacobi():
    """Solves the equation Ax=b via the Jacobi iterative method."""

    A=input("Enter initial matrix:(example input format:  [1, 2, 3], [4, 5, 6], [4, 5, 6] )")

    #convert A from string to numpy array
    A=array(ast.literal_eval(A))

    # Ensure non-zero diagonal elements
    while any(diag(A) == 0):
        for i in range(len(A)):
            if A[i, i] == 0:
                # Find a non-zero diagonal element in a lower row
                for j in range(i + 1, len(A)):
                    if A[j, j] != 0:
                        # Swap rows i and j
                        A[[i, j]] = A[[j, i]]
                        b[[i, j]] = b[[j, i]]
                        break  # Exit the inner loop after a successful swap

                # If no non-zero diagonal found in lower rows, swap with a column
                if A[i, i] == 0:
                    for j in range(i + 1, len(A)):
                        if A[i, j] != 0:
                            # Swap columns i and j
                            A[:, [i, j]] = A[:, [j, i]]
                            break  # Exit the inner loop after a successful swap

    b=input("Enter the right hand side of equation as a vector:")
    #convert b from string to numpy array
    b=array(ast.literal_eval(b))



    x=input("Enter initial values:")
    #convert x from string to numpy array
    x=array(ast.literal_eval(x))

    N=int(input("Enter the number of iterations:"))

    # Create an initial guess if needed
    if x is None:
        x = zeros(len(A[0]))



    # Create a vector of the diagonal elements of A
    # and subtract them from A
    D = zeros_like(A)#create the matrix with dimentions of A and fill with zeros
    fill_diagonal(D, A.diagonal())  # Fill diagonal of D with diagonal of A
    R = A - D



    #compute inverse of D
    D_inverse=linalg.inv(D)

    # Iterate for N times
    for i in range(N):
        x =dot(-dot(D_inverse,R),x)+dot(D_inverse,b)

    #print output
    print ("A:")
    pprint.pprint(A)

    print ("b:")
    pprint.pprint(b)
    print('\n')
    for i in range(len(x)):
      print(f"x{i+1}: {x[i]}")

# A = [2.0,-1,0],[-1.0,2.0,-1],[0,-1,2]
# b = [7.0,1.0,1.0]
# initial value = [0,0,0]
# Number of iterations: 14

jacobi()

```
